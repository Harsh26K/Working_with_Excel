### Survey Monkey File Transformation

#### Objectives:
* Make the header consistent [Done using Excel] 
* Add Column about how many people Answered particular Question
* Add Column about how many people responded with particular response for each question
* Rename some columns with short informative title

```python
# import required libraries
import numpy as np
import pandas as pd
```

```python
# import data 
dataset = pd.read_excel('Data - Survey Monkey Output.xlsx',sheet_name='new_worksheet')
dataset
```

```python
dataset.columns
```

1. Add Column about how many people Answered particular Question

```python
# Create a copy for callback in case of failures
work_data = dataset.copy()
work_data
```

we want to unpivot some columns. the melt() function is used.

```python
# The columns not to unpivot
demographic_columns = list(work_data.columns[:8])
demographic_columns
# The columns to unpivot
question_columns =list(work_data.columns[8:])
question_columns
```

```python
work_data_melted = work_data.melt(id_vars=demographic_columns,value_vars=question_columns,var_name='Question+Subquestion',value_name='Answer')
work_data_melted
```

```python
# Create copy for rollback
work_data_melted_copy = work_data_melted.copy()
work_data_melted_copy
```

we need to add columns "Question" and "Question+Subquestion" from another sheet to the main dataset. we will use pandas "merge()" function for this.  

```python
# Importing required sheet
questions_data = pd.read_excel('Data - Survey Monkey Output.xlsx',sheet_name='header')
questions_data
```

```python
questions_data_work = questions_data.copy()
```

```python
# removing unnecessary columns
questions_data_work.drop(columns=['Raw_Question','Subquestion'],inplace=True)
questions_data_work
```

```python
work_data_merged = pd.merge(left=work_data_melted_copy,right=questions_data_work,how="left",left_on="Question+Subquestion",right_on="Question+Subquestion")
# Check Before and After results to see if unnecessary merging happened
print("Before merge:",len(work_data_melted))
print('after merge:',len(work_data_merged))
work_data_merged
```

```python
# filtering NaN values to ease the calculation
respondants = work_data_merged[work_data_merged['Answer'].notna()]
respondants
```

```python
respondants.rename(columns={"Respondent ID ":"Id"},inplace=True)
```

```python
# to answer the given problem, we will count the unique Ids that responded to the particular question.
respondants = respondants.groupby("Question")["Id"].nunique().reset_index()
respondants = respondants.rename(columns={"Id":"total_response"})
respondants
```

```python
# we will add column "total response" to the main dataset using merge()
work_data_merged_two = pd.merge(left=work_data_merged,right=respondants,how="left",left_on="Question",right_on="Question")
print("Before merge:",len(work_data_merged))
print("After merge:",len(work_data_merged_two))
work_data_merged_two
```

2. Add Column about how many people responded with particular response for each question

```python
# to answer this, we will count the responses 
same_answer = work_data_merged.copy()
same_answer = same_answer.groupby(["Question+Subquestion","Answer"])["Respondent ID "].nunique().reset_index()
same_answer.rename(columns={"Respondent ID ":"total_count"},inplace=True)
same_answer
```

```python
work_data_merged_three = pd.merge(left=work_data_merged_two,right=same_answer,how="left",left_on=["Question+Subquestion","Answer"],right_on=["Question+Subquestion","Answer"])
print("Before merge:",len(work_data_merged_two))
print("After merge:",len(work_data_merged_three))
work_data_merged_three
```

```python
# filling NaN values with 0
work_data_merged_three["total_count"].fillna(0,inplace=True)
work_data_merged_three
```

```python
work_data_merged_three.columns
```

3. Rename some columns with short informative title

```python
work_data_merged_three.rename(columns={'Identify which division you work in. Response':"division primary",
'Identify which division you work in. Other (please specify)':"division secondary",
'Which of the following best describes your position level? Response':"position",
'Which generation are you apart of? Response':"generation",
'Please select the gender in which you identify. Response':"gender",
'Which duration range best aligns with your tenure at your company? Response':"tenure",
'Which of the following best describes your employment type? Response':"employment type"},inplace=True)

work_data_merged_three
```

```python
work_data_merged_three.to_csv("Modified_Survey_Monkey_Output.csv")
```
